```python
import numpy as np
#matplotlib is a library for data visualisation.
#typically in the form of plots,graphs and charts.
""" 
pyplot is an API (Application Programming Interface) for Python's
matplotlib that effectively makes matplotlib a viable source alternative
to MATLAB
"""
import matplotlib.image as mpimg #used for working with images in python
import matplotlib.pyplot as plt
# you can generate random numbers in python using random module.
"""
Python offers random module that can generate random numbers. These are
pseudo-random number as the sequence of number generated depends on the seed.
if the seeding  value is the the same, the sequence will be the same. 
"""
import random
# scikit-image or skimage- (for img preprocessing) is an open source python package designed for image preprocessing
"""
skimage includes algorithms for segmentation,geometric transformations,
color space manipulation, analysis filtering,morphology, feature detection,
and more
"""
# the exposure of an image tells whether the image is too light,to dark or balanced.
# exposure- for local contrast enhancement
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize

```

```python
img = mpimg.imread('priyanka.jpg')
```

```python
plt.imshow(img)
```

#Rescaling

```python
img_rescale = resize(img,(150,150))
```

```python
plt.imshow(img_rescale)
```

```python
img_rescale1 = resize(img,(50,50))
plt.imshow(img_rescale1)
```

#Flipping


```python
horizontal_flip = np.fliplr(img)
plt.imshow(horizontal_flip)
```

```python
vertical_flip=np.flipud(img)
plt.imshow(vertical_flip)
```

#Rotation of image

```python
from skimage import transform
transform_image = transform.rotate(img, random.uniform(-40,40))
plt.imshow(transform_image)
```

#Adding noise to image

```python
img_noise = random_noise(img, mode='s&p', clip = True)
plt.imshow(img_noise)
```
